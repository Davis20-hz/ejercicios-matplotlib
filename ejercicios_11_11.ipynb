```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0, 2*np.pi, 100)
y1=np.sin(x)
y2=np.cos(x)

plt.figure(figsize=(10,5))

plt.plot(x,y1,label='sin(x)')
plt.plot(x,y2,label='cos(x)')

plt.xlabel('x')
plt.ylabel("y")
plt.title("funciones trigonometricas", fontsize=22)
plt.xlabel("eje x", fontsize=20)
plt.ylabel("eje y", fontsize=20)

plt.yticks(fontsize=18)
plt.xticks(fontsize=18)
plt.grid(True)
plt.legend(loc='upper center', title= "func. trigonometricas")
plt.show()
```

```python
import random
import numpy as np
from scipy import stats

data=np.random.randint(0, 101, size=200)

media=np.mean(data)
mediana=np.median(data)
moda=stats.mode(data)
varianza=np.var(data)
desviacion_estandar=np.std(varianza)

print("media:", media)
print("mediana:", mediana)
print("moda:", moda)
print("varianza", varianza)
print("desviacion estandar:", desviacion_estandar)

plt.figure()
plt.hist(data, bins=10, edgecolor="red", linewidth=1.5, alpha=0.7)  # 10 bins (intervalos)
plt.title("Histograma de nmeros aleatorios (1100)")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()
```
