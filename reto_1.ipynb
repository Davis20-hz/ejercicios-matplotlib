```python
import random

def piedra_papel_tijeras():
    opciones = ["piedra", "papel", "tijeras"]
    computadora = random.choice(opciones)

    jugador = input("Elige piedra, papel o tijeras: ").lower()# .lowe= que no importe como se escriba (estandariza el texto para compararlo fácilmente.)

    if jugador not in opciones:
        print("Intenta de nuevo con una de las opciones")
        return

    print(f"Computadora eligió: {computadora}")

    if jugador == computadora:
        print("empate")
    elif (jugador == "piedra" and computadora == "tijeras") or \
         (jugador == "papel" and computadora == "piedra") or \
         (jugador == "tijeras" and computadora == "papel"):
        print("ganaste")
    else:
        print("Perdiste :(")


piedra_papel_tijeras()
```

```python
import random

print("juego de Dados ")

while True:
    dado = random.randint(1, 6)
    print(f"El dado cayó en: {dado}")

    if dado == 1:
        print("Cayó 1 → se vuelve a lanzar.")
        continue
    elif dado == 6:
        print("Cayó 6 → es empate")
        break
    elif dado in [2, 3]:
        print("Cayó 2 o 3 → gana Juan")
        break
    elif dado in [4, 5]:
        print("Cayó 4 o 5 → gana María")
        break
```
